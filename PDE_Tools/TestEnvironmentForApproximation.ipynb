```
def function(t,y): #Create a derivative function
  return (y* 2 + y)
```

```
def EulerMethod(t_0,y_0,delta_t,end_interval):  #function return array of y and t
  n = int((end_interval - t_0)/delta_t) + 1 #number of point
  y = np.zeros([n])                          #Define pre-values of y
  t = np.zeros([n])                          #Define pre-values of t
  t[0] = t_0                                 #t[0] is the value of t_0
  y[0] = y_0                                 #y[0] is the value of y_0
  for i in range(1,n):                       #for-loop to calculate the approximation by Euler method
    y[i] = delta_t * (function(t_0,y[i-1])) + y[i-1] #Using function y(n) = delta*t * derivative(t(n-1) and y(n-1)) + y(n-1)
    t_0 = t_0 + delta_t
    t[i] = t_0
  #Plot the approximation
  plt.plot(t,y,'-b')
  plt.xlabel("Value of t")
  plt.ylabel("Value of y")
  plt.title("Approximate Solution With Euler's Method")
  plt.show()
  return y,t
```

```
def RK2method(t_0,y_0,delta_t,end_interval):
    n = int((end_interval - t_0)/delta_t) + 1 #number of point
    y = np.zeros([n])                         #Define pre-values of y
    t = np.zeros([n])                         #Defi"ne pre-values of t
    t[0] = t_0                                #t[0] is the value of t_0
    y[0] = y_0                                #y[0] is the value of y_0
    for i in range(1,n):                      #for-loop to calculate the approximation by Runge Kutta
      k1 = delta_t * function(t_0,y[i-1])
      k2 = delta_t * function(t_0 + 0.5 * delta_t, y[i-1] + 0.5 * k1) 
      y[i] = y[i-1] + (k1+k2)/2 
      t_0 = t_0 + delta_t
      t[i] = t_0
    #Plot the approximation
    plt.plot(t,y,'-b')
    plt.xlabel("Value of t")
    plt.ylabel("Value of y")
    plt.title("Approximate Solution With RK2's Method")
    plt.show()
    return y,t
```

```
def RK4method(t_0,y_0,delta_t,end_interval):
    n = int((end_interval - t_0)/delta_t) + 1 #number of point
    y = np.zeros([n])                         #Define pre-values of y
    t = np.zeros([n])                         #Define pre-values of t
    t[0] = t_0                                #t[0] is the value of t_0
    y[0] = y_0                                #y[0] is the value of y_0
    for i in range(1,n):                      #for-loop to calculate the approximation by Runge Kutta
      k1 = delta_t * function(t_0,y[i-1])
      k2 = delta_t * function(t_0 + 0.5 * delta_t, y[i-1] + 0.5 * k1) 
      k3 = delta_t * function(t_0 + 0.5 * delta_t, y[i-1] + 0.5 * k2) 
      k4 = delta_t * function(t_0 + delta_t,y[i-1] + k3)
      y[i] = y[i-1] + (1.0 / 6.0)*(k1 + 2 * k2 + 2 * k3 + k4) 
      t_0 = t_0 + delta_t
      t[i] = t_0
    #Plot the approximation
    plt.plot(t,y,'-b')
    plt.xlabel("Value of t")
    plt.ylabel("Value of y")
    plt.title("Approximate Solution With RK4's Method")
    plt.show()
    return y,t
```

```
t_0 = 0
y_0 = 1
delta_t = 0.1
end_interval = 2
y_el, t_el = EulerMethod(t_0,y_0,delta_t,end_interval)
y_rk2, t_rk2 = RK2method(t_0,y_0,delta_t,end_interval)
y_rk4, t_rk4 = RK4method(t_0,y_0,delta_t,end_interval)
plt.plot(t_el,y_el,'-g',label ='Euler Method',linestyle = 'dashed')
plt.plot(t_rk2,y_rk2,'-r', label = 'RK2 Method', linestyle = 'dashed')
plt.plot(t_rk4,y_rk4,'-b', label = 'RK4 Method')
plt.xlabel("Value of t")
plt.ylabel("Value of y")
plt.title("Approximate Solution With All Methods")
plt.legend()
```
